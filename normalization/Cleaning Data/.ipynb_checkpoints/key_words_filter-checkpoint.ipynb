```python
import json, csv, string, re, emoji
import pandas as pd
import numpy  as np
#PNL
from nltk.tokenize import TweetTokenizer
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('out_dataset.csv')
#data = pd.read_csv('data_set.csv')
data.dropna(inplace = True)
```

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

def remove_dots(text):
    return  re.compile(
        f'[{re.escape(string.punctuation)}]'
    ).sub(' ', text)

def remove_stopwords(text):
    make   = TweetTokenizer()
    tokens = make.tokenize( remove_dots( remove_numbers(text) ) )
    return [w.lower() for w in tokens if not w in set(stopwords.words('english'))]
```

```python
words = set(remove_stopwords(
    " ".join([x for x in open('disasters_words.txt')])
))
```

```python
words
```

```python
def find_emojis(lista):
    allchars = [str for str in lista.replace(" ","")]
    return [c for c in allchars if c in emoji.UNICODE_EMOJI] 

def normalize_text():
    
    #Cria uma dataSerie para ser adicionado ao dataFrame
    data['interest'] = False
    
    for index, line in enumerate(data['text']):
        TOKENS = remove_stopwords(line)
        data.at[index, 'text'] = " ".join(TOKENS)
        for token in TOKENS:
            if token in words:
                data.at[index, 'interest'] =  True
normalize_text()
```
